## 2. Задание. Синтаксис. Сложение четных чисел

Написать скрипт, который считывает с клавиатуры числа, складывает только четные и выводит результат на экран. Правильность решения можете проверить в контесте в задаче Синтаксис. Сложение четных чисел.

#### Из контеста
На вход с клавиатуры подаются числа (по одному). Нужно вывести на экран сумму всех четных чисел. Ввод прекращается, если введена пустая строка (""). При любой ситауации, когда не попадаются четные числа, вывести на экран нужно 0 (например, числа не были введены вообще или были введены только нечетные числа).

#### Примечания
Ответ проверяется для целочисенного типа данных посимвольно. Т.е., допустим, ответ 100 будет правильным, в то время как 100.0 выдаст ошибку, поскольку он отличается от правильного на два символа .0

```python
# Решение 1, не подходит для контеста
s = input("Напишите снизу ряд чисел:\n")

punctuation = ['.', ',', ':', ';', '!', '?', '(', ')']

word_list = s.split()

for i, word in enumerate(word_list):
    if word[-1] in punctuation:
        word_list[i] = word[:-1]
        word = word_list[i]
    if word[0] in punctuation:
        word_list[i] = word[1:]

new_lst = [i for i in map(int,word_list) if i % 2 == 0]
a = sum(new_lst)
print('Суммма четных чисел')
print(a)
```

## 7. Задание.Коллекции. Множества.

С клавиатуры строка, содержащая произвольное количество слов через запятую и пробел. Слова могут повторяться. Нужно вывести на экран все уникальные слова в алфавитном порядке, также через пробел и запятую. Оценить правильность решения вы можете в контесте в задаче Коллекции. Множества

#### Из контеста
С клавиатуры вводится строка, содержащая произвольное количество слов через запятую и пробел. Слова могут повторяться. Нужно вывести на экран все слова в алфавитном порядке по одному разу, также через пробел и запятую.

#### Примечание
Обратите внимание, что в примере у первого слова "воробей" стоит запятая, а у второго - нет. В решении такие случае следует учесть. Кроме того, слова могут быть написаны в разном регистре. Вывести их нужно в нижнем регистре. Одинаковые слова, написанные в разных регистрах, считаем одинаковыми.

```python
s = input()

punctuation = ['.', ',', ':', ';', '!', '?', '(', ')']

word_list = s.lower().split()

for i, word in enumerate(word_list):
    if word[-1] in punctuation:
        word_list[i] = word[:-1]
        word = word_list[i]
    if word[0] in punctuation:
        word_list[i] = word[1:]

lst_set = set(word_list)    
lst_sort = sorted(list(lst_set))
last_lst = ', '.join(lst_sort)
print(last_lst)
```

## 8. Задание. Коллекции. Словари.

С клавиатуры вводятся слова через запятую с пробелом. Выведите на экран три наиболее часто встречаемых слова, вместе с количеством этих слов. Количество должно быть отделено от слова двоеточием и пробелом. Каждая пара слово-количество должна быть выведена на отдельной строчке. Для простоты гарантируется, что в строке нет слов с одинаковой встречаемостью. Проверить свое решение вы можете в контесте в задаче Коллекции. Словари

##### Из контеста
С клавиатуры вводятся слова через запятую с пробелом. Выведите на экран три наиболее часто встречаемых слова, вместе с количеством этих слов. Количество должно быть отделено от слова двоеточием и пробелом. Каждая пара слово-количество должна быть выведена на отдельной строчке. Для простоты гарантируется, что в строке нет слов с одинаковой встречаемостью.

##### Примечание
Рекомендация: чтобы быстрее решить задачу, нужно составить словарь, в котором ключами будут встреченные слова, а значениями - их количество в строке. Далее работайте с этим словарем по своему усмотрению. Далее один из вариантов - отсортировать кортежи, в которых на первом месте стоит количество слов, а на втором - сами слова.

Учтите, что различных слов в строке может быть меньше трех. В этом случае нужно вывести все.

```python
spisok = input()
spisok = spisok.strip().lower().split(', ')
unique_spisok = list(set(spisok))
spisok_dict = dict()
lst=[]
for i in range(len(unique_spisok)):
    spisok_dict.update({unique_spisok[i]:spisok.count(unique_spisok[i])})
for key, value in spisok_dict.items():
    lst += [[str(value),key]]
lst = sorted(lst, reverse = True)
for i in range(len(lst)):
    lst[i] = ': '.join(lst[i][::-1])
print(' '.join(lst[:3]))
```

## 9. Задание. Даты.

С клавиатуры вводится дата в формате DD-MM-YYYY. Нужно вывести дату начала недели, к которой относится введенная дата (дата понедельника недели), в таком же формате. Проверить свое решение вы можете в контесте в задаче Даты.

#### Из контеста
С клавиатуры вводится дата в формате DD-MM-YYYY. Нужно вывести дату начала недели, к которой относится введенная дата (дата понедельника недели), в таком же формате.

#### Примечание
Если введен понедельник - нужно вывести его же.

```python
import datetime
str1 = input()
data1 = datetime.datetime.strptime(str1, "%d-%m-%Y")
delta = datetime.timedelta(days=data1.weekday())
data2 = data1 - delta
print("{}-{}-{}".format(str(data2.day).zfill(2), str(data2.month).zfill(2), data2.year))
```
